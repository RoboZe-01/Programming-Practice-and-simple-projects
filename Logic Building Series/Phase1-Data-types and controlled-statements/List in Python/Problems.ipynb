### **Easy Problems (Fundamentals)**  
1. **Sum of List**  
   - Given a list of numbers, return the sum.  
   - Example: `[1, 2, 3]` → `6`  

2. **Find Maximum/Minimum**  
   - Return the largest/smallest number in a list.  
   - Example: `[4, 2, 9]` → `9` (max), `2` (min)  

3. **Count Occurrences**  
   - Count how many times a given value appears in a list.  
   - Example: `[1, 2, 2, 3]`, target=`2` → `2`  

4. **Remove Duplicates**  
   - Return a new list with duplicates removed (order preserved).  
   - Example: `[1, 2, 2, 3]` → `[1, 2, 3]`  

5. **Merge Two Lists**  
   - Concatenate two lists into one.  
   - Example: `[1, 2]` + `[3, 4]` → `[1, 2, 3, 4]`  

6. **List Reversal**  
   - Reverse a list without using built-in `reverse()`.  
   - Example: `[1, 2, 3]` → `[3, 2, 1]`  

7. **Even/Odd Filter**  
   - Separate even and odd numbers into two lists.  
   - Example: `[1, 2, 3, 4]` → `[2, 4]`, `[1, 3]`  

8. **Flatten a Nested List**  
   - Convert a nested list into a flat list (1-level deep).  
   - Example: `[[1, 2], [3, 4]]` → `[1, 2, 3, 4]`

### **Intermediate Problems (Logic & Algorithms)**  
9. **Two Sum**  
   - Find two numbers in a list that add up to a target.  
   - Example: `[2, 7, 11, 15]`, target=`9` → `[2, 7]`  

10. **Rotate List**  
    - Rotate a list right by `k` steps.  
    - Example: `[1, 2, 3, 4]`, `k=2` → `[3, 4, 1, 2]`  

11. **Find Missing Number**  
    - Given a list of numbers from `1` to `n` with one missing, find it.  
    - Example: `[1, 3, 4]` → `2`  

12. **List Chunking**  
    - Split a list into smaller lists of size `n`.  
    - Example: `[1, 2, 3, 4]`, `n=2` → `[[1, 2], [3, 4]]`  

13. **Move Zeroes to End**  
    - Move all zeroes to the end while preserving order.  
    - Example: `[0, 1, 0, 3]` → `[1, 3, 0, 0]`  

14. **Longest Consecutive Sequence**  
    - Find the length of the longest consecutive sequence.  
    - Example: `[100, 4, 200, 1, 3, 2]` → `4` (sequence: `1, 2, 3, 4`)  

15. **Merge Intervals**  
    - Merge overlapping intervals in a list of `[start, end]` pairs.  
    - Example: `[[1, 3], [2, 6]]` → `[[1, 6]]`  

16. **Find Duplicates**  
    - Return all duplicate elements in a list.  
    - Example: `[1, 2, 2, 3, 3]` → `[2, 3]`  

17. **List Product Except Self**  
    - Return a list where each element is the product of all others except itself.  
    - Example: `[1, 2, 3]` → `[6, 3, 2]`

### **Challenging (Real-World Applications)**  
18. **Pascal’s Triangle**  
    - Generate the first `n` rows of Pascal’s Triangle.  
    - Example: `n=3` → `[[1], [1, 1], [1, 2, 1]]`  

19. **Spiral Matrix**  
    - Traverse a 2D list in spiral order.  
    - Example: `[[1, 2, 3], [4, 5, 6]]` → `[1, 2, 3, 6, 5, 4]`  

20. **Leaderboard Update**  
    - Given a leaderboard (sorted list) and a new score, insert it in the correct position.  
    - Example: `[100, 90, 80]`, score=`95` → `[100, 95, 90, 80]`  

21. **Custom Sort Order**  
    - Sort a list based on another list’s order.  
    - Example: `['a', 'b', 'c']`, order=`['b', 'a', 'c']` → `['b', 'a', 'c']`  

22. **Frequency Sort**  
    - Sort a list by frequency of elements (ascending/descending).  
    - Example: `[4, 4, 2, 1, 1, 1]` → `[2, 4, 4, 1, 1, 1]` (ascending frequency)

- **Algorithms**: Two-pointer, sliding window, sorting.  
- **Real-World**: Matrix traversal, leaderboards, frequency analysis.  

**Tip**: Start with easy problems, then gradually tackle intermediate ones. Use list comprehensions for concise solutions!
