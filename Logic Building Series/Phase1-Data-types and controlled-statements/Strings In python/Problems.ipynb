### **Easy Problems (Basic String Manipulation)**  
1. **Reverse a String**  
   - Given a string, return it reversed.  
   - Example: `"hello"` → `"olleh"`  

2. **Check Palindrome**  
   - Determine if a string reads the same backward (case-sensitive).  
   - Example: `"madam"` → `True`, `"hello"` → `False`  

3. **Count Vowels**  
   - Count the number of vowels (`a, e, i, o, u`) in a string.  
   - Example: `"apple"` → `2`  

4. **Remove Whitespace**  
   - Remove all leading/trailing spaces from a string.  
   - Example: `"  hello  "` → `"hello"`  

5. **Capitalize First Letter**  
   - Capitalize the first letter of a string.  
   - Example: `"python"` → `"Python"`  

6. **Count Words in a Sentence**  
   - Return the number of words in a sentence (separated by spaces).  
   - Example: `"Hello world"` → `2`  

7. **Find the Longest Word**  
   - Given a sentence, return the longest word.  
   - Example: `"I love Python"` → `"Python"`  

8. **Replace Substring**  
   - Replace all occurrences of a substring with another.  
   - Example: `"Hello world"`, `"world"` → `"Python"` → `"Hello Python"`

### **Medium Problems (Advanced String Operations)**  
9. **Check Anagram**  
   - Determine if two strings are anagrams (case-insensitive).  
   - Example: `"listen"`, `"silent"` → `True`  

10. **String Compression**  
    - Compress consecutive duplicate characters (case-sensitive).  
    - Example: `"aaabbbcc"` → `"a3b3c2"`  

11. **First Unique Character**  
    - Find the index of the first non-repeating character.  
    - Example: `"leetcode"` → `0` (first `'l'` is unique)  

12. **Valid Parentheses**  
    - Check if a string has balanced parentheses `()`, `[]`, `{}`.  
    - Example: `"({[]})"` → `True`, `"([)]"` → `False`  

13. **String to Integer (atoi)**  
    - Convert a string to an integer (handle signs and overflow).  
    - Example: `" -42"` → `-42`  

14. **Longest Substring Without Repeating Characters**  
    - Return the length of the longest substring with unique characters.  
    - Example: `"abcabcbb"` → `3` (`"abc"`)  

15. **Group Anagrams**  
    - Given a list of strings, group anagrams together.  
    - Example: `["eat", "tea", "tan", "ate", "nat", "bat"]` → `[["eat","tea","ate"], ["tan","nat"], ["bat"]]`  

16. **Longest Palindromic Substring**  
    - Find the longest substring that is a palindrome.  
    - Example: `"babad"` → `"bab"` or `"aba"`

### **Bonus (Real-World Applications)**  
17. **Extract Domain from Email**  
    - Given an email, extract the domain.  
    - Example: `"user@example.com"` → `"example.com"`  

18. **Mask Email Address**  
    - Mask the middle characters of an email (except first and last).  
    - Example: `"test@example.com"` → `"t**t@example.com"`  

19. **URL Slug Generator**  
    - Convert a title to a URL-friendly slug (lowercase, hyphens).  
    - Example: `"Hello World!"` → `"hello-world"`  

20. **Caesar Cipher Encoder**  
    - Shift each letter in a string by `n` positions (wrap around).  
    - Example: `"abc"`, `n=3` → `"def"`
